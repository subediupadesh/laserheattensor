```python
def Boiling_Melting(FCC=[0.28, 0.18, 0.22, 0.32], LIQ=[0.35, 0.13, 0.15, 0.37]):
    Boiling ={"Co": 3143, "Cr": 2944, "Fe": 3135, "Ni": 3186}
    Melting ={"Co": 1768, "Cr": 2130, "Fe": 1808, "Ni": 1726}

    Melting_T = Melting["Co"]*FCC[0] + Melting["Cr"]*FCC[1] + Melting["Fe"]*FCC[2] + Melting["Ni"]*FCC[3]
    Boiling_T = Boiling["Co"]*LIQ[0] + Boiling["Cr"]*LIQ[1] + Boiling["Fe"]*LIQ[2] + Boiling["Ni"]*LIQ[3]

    return Melting_T, Boiling_T
```

```python
import numpy as np

def get_max_meltpool_depth_y(eta_liq, no_melt_value=np.nan):
    """
    Find maximum meltpool depth y-coordinate for each time step.

    Parameters
    ----------
    eta_liq : np.ndarray
        Binary ETALIQ array with shape (Nt, Ny, Nx).
        0 = FCC, 1 = LIQUID.

    no_melt_value : int or float
        Value returned when no meltpool exists at a time step.

    Returns
    -------
    max_depth_y : np.ndarray
        Array of shape (Nt,) containing the y-coordinate where the
        deepest meltpool point first appears when scanning from bottom
        Ny-1 to top 0.
    """

    if eta_liq.ndim != 3:
        raise ValueError(f"Expected eta_liq shape (Nt, Ny, Nx), got {eta_liq.shape}")

    Nt, Ny, Nx = eta_liq.shape

    # True if a given y-row contains at least one meltpool pixel over all x
    row_has_liquid = np.any(eta_liq == 1, axis=2)   # shape: (Nt, Ny)

    # Reverse y-axis so scan starts from bottom: y = Ny-1, Ny-2, ..., 0
    row_has_liquid_reversed = row_has_liquid[:, ::-1]

    # Check which time steps actually contain liquid
    has_melt = np.any(row_has_liquid_reversed, axis=1)

    # First True in reversed y-direction
    first_liquid_from_bottom = np.argmax(row_has_liquid_reversed, axis=1)

    # Convert reversed index back to original y-coordinate
    max_depth_y = (Ny - 1) - first_liquid_from_bottom

    # Assign no_melt_value where there is no liquid at all
    max_depth_y = max_depth_y.astype(float)
    max_depth_y[~has_melt] = no_melt_value

    return max_depth_y+1

ETA_D_P300S45cTF3 = get_max_meltpool_depth_y(P300S45cTF3)
ETA_D_P400S45cTF3 = get_max_meltpool_depth_y(P400S45cTF3)
ETA_D_P500S45cTF3 = get_max_meltpool_depth_y(P500S45cTF3)

ETA_D_P300S70cTF3 = get_max_meltpool_depth_y(P300S70cTF3)
ETA_D_P400S70cTF3 = get_max_meltpool_depth_y(P400S70cTF3)
ETA_D_P500S70cTF3 = get_max_meltpool_depth_y(P500S70cTF3)


ETA_D_P300S45cTF4 = get_max_meltpool_depth_y(P300S45cTF4)
ETA_D_P400S45cTF4 = get_max_meltpool_depth_y(P400S45cTF4)
ETA_D_P500S45cTF4 = get_max_meltpool_depth_y(P500S45cTF4)

ETA_D_P300S70cTF4 = get_max_meltpool_depth_y(P300S70cTF4)
ETA_D_P400S70cTF4 = get_max_meltpool_depth_y(P400S70cTF4)
ETA_D_P500S70cTF4 = get_max_meltpool_depth_y(P500S70cTF4)
```
