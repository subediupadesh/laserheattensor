```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({"font.family": "Play",})
```

```python
path = os.path.abspath(f'')
```

## LIQUID

```python
df = pd.read_csv("LIQUID_factor_matrices_R6_wide.csv")
d_Co = df[(df["variable"] == "Co") & (df["matrix"] == "A")]
d_Cr = df[(df["variable"] == "Cr") & (df["matrix"] == "B")]
d_Fe = df[(df["variable"] == "Fe") & (df["matrix"] == "C")]
d_T  = df[(df["variable"] == "Temperature")  & (df["matrix"] == "D")]
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=[12, 8], frameon=True)
ax1.set_facecolor((0, 0, 0, 0))

ax1.plot(d_Co["x_value"], d_Co["r=1"], lw=3, marker="o", ms=8, ls='-', color='tab:red', label="Rank 1")
ax1.plot(d_Co["x_value"], d_Co["r=2"], lw=3, marker="s", ms=8, ls='-', color='tab:green', label="Rank 2")
ax1.plot(d_Co["x_value"], d_Co["r=3"], lw=3, marker="^", ms=8, ls='-', color='blue', label="Rank 3")
ax1.plot(d_Co["x_value"], d_Co["r=4"], lw=3, marker="D", ms=8, ls='-', color='tab:orange', label="Rank 4")
ax1.plot(d_Co["x_value"], d_Co["r=5"], lw=3, marker="P", ms=8, ls='-', color='magenta', label="Rank 5")
ax1.plot(d_Co["x_value"], d_Co["r=6"], lw=3, marker="X", ms=8, ls='-', color='olive', label="Rank 6")


ax1.set_xlim(-0.02, 0.42)
ax1.set_ylim(-0.45, 0.45)
ax1.set_xticks([0.0, 0.1, 0.2, 0.3, 0.4], labels=['0.0', '0.1', '0.2', '0.3', '0.4'], size=35, weight='bold', color='k', rotation=0) 
ax1.set_yticks([-0.4, -0.2, 0.0, 0.2, 0.4], labels=['-0.4', '-0.2', '0.0', '0.2', '0.4'], size=35, weight='bold', color='k', rotation=0) 
ax1.tick_params(axis="both", width=5, length=15, direction='inout')
ax1.spines[:].set_linewidth(4)
ax1.grid(True)

plt.savefig("../figures/Co_Factor_LIQ.png", transparent=True, bbox_inches="tight", pad_inches=0.12)
plt.show()
```
